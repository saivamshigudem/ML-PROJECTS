```python
from sklearn.preprocessing import StandardScaler
```

```python
print("\n")

print("="*70)
print("Missing Values Before Treatment")
print("="*70)

print(df.isnull().sum())
```

```python
for column in df.columns:

    if df[column].dtype == "object":

        df[column].fillna(

            df[column].mode()[0],

            inplace=True

        )

    else:

        df[column].fillna(

            df[column].median(),

            inplace=True

        )

print("\n")

print("="*70)
print("Missing Values After Treatment")
print("="*70)

print(df.isnull().sum())
```

```python
X = df.drop(

    "Class",

    axis=1

)

y = df["Class"]

print("\n")

print("="*70)
print("Feature Matrix and Target Variable")
print("="*70)

print("Feature Shape :", X.shape)

print("Target Shape  :", y.shape)
```

```python
print("\n")

print("="*70)
print("Feature Matrix")
print("="*70)

display(X.head())
```

```python
print("\n")

print("="*70)
print("Target Variable")
print("="*70)

display(y.head())
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(

    X

)

print("\n")

print("="*70)
print("Feature Scaling Completed")
print("="*70)

print("Scaled Data Shape :", X_scaled.shape)
```

```python
X_scaled_df = pd.DataFrame(

    X_scaled,

    columns=X.columns

)

print("\n")

print("="*70)
print("Scaled Feature Matrix")
print("="*70)

display(X_scaled_df.head())
```

```python
print("\n")

print("="*70)
print("Mean of Scaled Features")
print("="*70)

display(

    X_scaled_df.mean().round(2)

)
```

```python
print("\n")

print("="*70)
print("Standard Deviation of Scaled Features")
print("="*70)

display(

    X_scaled_df.std().round(2)

)
```

```python
from sklearn.cluster import DBSCAN

from sklearn.decomposition import PCA

from sklearn.metrics import silhouette_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
dbscan = DBSCAN(

    eps=2,

    min_samples=10,

    metric="euclidean"

)
```

```python
cluster_labels = dbscan.fit_predict(

    X_scaled

)

print("\n")

print("="*70)
print("DBSCAN Model Trained Successfully")
print("="*70)
```

```python
df["Cluster"] = cluster_labels

print("\n")

print("="*70)
print("Cluster Labels Added")
print("="*70)

display(df.head())
```

```python
print("\n")

print("="*70)
print("Cluster Distribution")
print("="*70)

print(

    df["Cluster"]

    .value_counts()

    .sort_index()

)
```

```python
df["Anomaly"] = np.where(

    df["Cluster"] == -1,

    1,

    0

)

print("\n")

print("="*70)
print("Anomaly Detection Summary")
print("="*70)

print(

    df["Anomaly"]

    .value_counts()

)
```

```python
total_anomalies = df["Anomaly"].sum()

print("\n")

print("="*70)
print("Total Detected Anomalies")
print("="*70)

print("Detected Anomalies :", total_anomalies)

print("Normal Records     :", len(df)-total_anomalies)
```

```python
n_clusters = len(

    set(cluster_labels)

) - (1 if -1 in cluster_labels else 0)

print("\n")

print("="*70)
print("Number of Clusters")
print("="*70)

print("Clusters Formed :", n_clusters)
```

```python
import numpy as np
from sklearn.metrics import silhouette_score

print("\n")
print("="*70)
print("Silhouette Score (Sampled)")
print("="*70)

if n_clusters > 1:
    # Sample up to 2000 points (or fewer if dataset is smaller)
    sample_size = min(2000, len(X_scaled))
    sample_idx = np.random.choice(len(X_scaled), size=sample_size, replace=False)

    score = silhouette_score(
        X_scaled[sample_idx],
        cluster_labels[sample_idx]
    )

    print("Silhouette Score (approx):", round(score, 4))
else:
    score = None
    print("Silhouette Score cannot be calculated.")
```

```python
pca = PCA(

    n_components=2,

    random_state=42

)

X_pca = pca.fit_transform(

    X_scaled

)

print("\n")

print("="*70)
print("PCA Transformation Completed")
print("="*70)
```

```python
plt.figure(figsize=(10,7))

plt.scatter(

    X_pca[:,0],

    X_pca[:,1],

    c=cluster_labels,

    cmap="viridis",

    s=10

)

plt.title("DBSCAN Clustering")

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.colorbar(label="Cluster")

plt.show()
```

```python
plt.figure(figsize=(10,7))

plt.scatter(

    X_pca[:,0],

    X_pca[:,1],

    color="lightgray",

    s=10,

    label="Normal"

)

plt.scatter(

    X_pca[df["Anomaly"]==1,0],

    X_pca[df["Anomaly"]==1,1],

    color="red",

    s=25,

    label="Anomaly"

)

plt.title("Detected Anomalies")

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.legend()

plt.show()
```

```python
print("\n")

print("="*70)
print("Comparison with Actual Labels")
print("="*70)

cm = confusion_matrix(

    y,

    df["Anomaly"]

)

print("Confusion Matrix")

print(cm)
```

```python
print("\n")

print("="*70)
print("Classification Report")
print("="*70)

print(

    classification_report(

        y,

        df["Anomaly"],

        zero_division=0

    )

)
```

```python
import joblib

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(

    n_neighbors=5

)

knn_model.fit(

    X_scaled,

    df["Anomaly"]

)

print("\n")

print("="*70)

print("KNN Model Trained Successfully")

print("="*70)
```

```python
new_transaction = X.iloc[[0]].copy()

new_transaction_scaled = scaler.transform(

    new_transaction

)

prediction = knn_model.predict(

    new_transaction_scaled

)

print("\n")

print("="*70)

print("Prediction for New Transaction")

print("="*70)

if prediction[0] == 1:

    print("Prediction : Anomaly Transaction")

else:

    print("Prediction : Normal Transaction")
```

```python
probability = knn_model.predict_proba(

    new_transaction_scaled

)

print("\n")

print("="*70)

print("Prediction Probability")

print("="*70)

print(probability)
```
